# 이미지 처리

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read image
image = cv2.imread('./data/car.png',1)
car_image= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
car_image_gray = cv2.imread('./data/car.png', 0)

plt.imshow(car_image_gray,'gray')
plt.show()
```


## Blur

### Blur

```python
blurred_1 = np.hstack([cv2.blur(car_image_gray,(3,3)),
                       cv2.blur(car_image_gray,(5,5)),
                       cv2.blur(car_image_gray,(9,9))])

plt.imshow(blurred_1, 'gray')
plt.show()
```

### Guassian fliter

```python
gaussia_blurred_1 = np.hstack([
    cv2.GaussianBlur(car_image_gray,(3,3),0),
    cv2.GaussianBlur(car_image_gray,(5,5),0),
    cv2.GaussianBlur(car_image_gray,(9,9),0),
])

plt.imshow(gaussia_blurred_1, 'gray')
plt.show()
```

## Sharpening

### Gaussian

```python
def draw(img, str):
    plt.imshow(img)
    plt.title(str)
    plt.show()
```

```python
#sharpening filter
filter = np.array([[-1,-1,-1],
                   [-1,-9,-1],
                   [-1,-1,-1]])

sharpen_img = cv2.filter2D(car_image,-1, filter)

draw(sharpen_img,'sharpen')
```
