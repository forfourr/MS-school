# 조명조건에 따른 이미지 보정

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#이미지 그리기
def draw(img, text):
    plt.imshow(img)
    plt.title(text)
    plt.show()
```

```python
#이미지 읽기
image = cv2.imread('../data/dark.png')
#'C:/Users/iiile/Vscode_jupyter/data/dark.png'


#RGB변경
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#이미지 밝기 조정
gamma = 0.5
img_corrected = cv2.pow(image/100.0, gamma)
ima_corrected = img_corrected*100.0

#원본과 비교
draw(image,'original')
draw(img_corrected,'brightness corrected')
```
