```python
#이미지를 늘리지거나 자르지 않고 정사각형으로 만드는 방법
#넓이와 높이 중 짧은 길이를 긴 길이에 맞추어 빈 곳을 채운다.


import os
import cv2
import matplotlib.pyplot as plt
import PIL
```

```python
def draw(img, text):
    plt.imshow(img)
    plt.title(text)
    plt.show()
```

## 내가 한 방법 resize -> concat

```python
img_path = '../data/dark.png'
image = cv2.imread(img_path)
h, w, c = image.shape
print(h)


side_img = cv2.imread('../data/orange.jpg')
print(side_img.shape)
#cropped_img = cv2.resize(side_img, (607, 133))
cropped_img = cv2.resize(side_img, (w, (w-h)//2))
draw(cropped_img,'crop')

result_img01 = cv2.vconcat([cropped_img, image])
result_img02 = cv2.vconcat([result_img01, cropped_img])
draw(result_img02,'result')


```

```python
image = cv2.imread('../data/face2.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
h, w, c = image.shape
print(image.shape)

side_img = cv2.imread('../data/orange.jpg')
cropped_img = cv2.resize(side_img, ((h-w)//2,h))

result_img01 = cv2.hconcat([cropped_img, image])
result_img02 = cv2.hconcat([result_img01, cropped_img])
draw(result_img02,'result')
```

```python
from PIL import Image
import numpy as np
```

```python
def size(pil_img, background_color):
    w, h= pil_img.size

    if w == h:
        return pil_img
    elif w > h:
        result = Image.new(pil_img.mode, (w,w), background_color)
        #draw(result,'01')
        result.paste(pil_img, (0,(w-h) //2))
        print((w-h) //2)
        #draw(result,'02')
        return result
    else:
        result = Image.new(pil_img.mode, (h,h), background_color)
        #draw(result,'01')
        result.paste(pil_img, (0,(h-w) //2))
        #draw(result,'02')
        return result

img = Image.open('../data/dark.png')
size(img,(255,0,0))
```
